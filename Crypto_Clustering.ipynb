```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv("Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaled_data = StandardScaler().fit_transform(
    df_market_data[["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d", "price_change_percentage_30d",
                    "price_change_percentage_60d", "price_change_percentage_200d", "price_change_percentage_1y"]]
)
```

```python
# Create a DataFrame with the scaled data
scaled_market_data = pd.DataFrame(scaled_data, columns=["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d", "price_change_percentage_30d",
                    "price_change_percentage_60d", "price_change_percentage_200d", "price_change_percentage_1y"])

# Copy the crypto names from the original data
scaled_market_data["coin_id"] = df_market_data.index

# Set the coinid column as index
scaled_market_data = scaled_market_data.set_index("coin_id")


# Display sample data
scaled_market_data.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,12))
k
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for i in k:
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
    model = KMeans(n_clusters=i, random_state=1)
# 2. Fit the model to the data using `scaled_market_data_df`
    model.fit(scaled_market_data)
# 3. Append the model.inertia_ to the inertia list
    inertia.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_curve = {"k": k, "inertia": inertia}
elbow_curve_df = pd.DataFrame(elbow_curve)
# Create a DataFrame with the data to plot the Elbow curve
elbow_curve_df.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_curve_plot = elbow_curve_df.hvplot(
    x = "k",
    y = "inertia", 
    xticks = k
)
elbow_curve_plot
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for 'k' is 4. 

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4)
```

```python
# Fit the K-Means model using the scaled data
model.fit(scaled_market_data)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
cluster_predict = model.predict(scaled_market_data)

# Print the resulting array of cluster values.
cluster_predict
```

```python
# Create a copy of the DataFrame
cluster_predict_df = scaled_market_data.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
cluster_predict_df["cluster_number"] = cluster_predict

# Display sample data
cluster_predict_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
clusters_scatter_plot = cluster_predict_df.hvplot.scatter(
    x = "price_change_percentage_24h", 
    y = "price_change_percentage_7d", 
    by = "cluster_number",
    hover_cols = ['coin_id']
)
clusters_scatter_plot
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components = 3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
no_clusters = cluster_predict_df.drop(columns="cluster_number")
pca_model = pca.fit_transform(scaled_market_data)

# View the first five rows of the DataFrame. 
pca_model[0:5]

```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

```python
# Finding the total explained variance 
pca.explained_variance_ratio_.sum()
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 89.5%

```python
# Create a new DataFrame with the PCA data.

# Creating a DataFrame with the PCA data

pca_data = pd.DataFrame(pca_model, columns=["PC1", "PC2", "PC3"])

# Copy the crypto names from the original data
pca_data["coin_id"] = df_market_data.index

# Set the coinid column as index
pca_data = pca_data.set_index("coin_id")

# Display sample data
pca_data.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,12))
k
```

```python
# Create an empty list to store the inertia values
inertia_pca = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `pca_data`
# 3. Append the model.inertia_ to the inertia list
for i in k: 
    model_pca = KMeans(n_clusters=i, random_state=1)
    model_pca.fit(pca_data)
    inertia_pca.append(model_pca.inertia_)
    
inertia_pca
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_inertia_pca = {'k': k, 'inertia_pca': inertia_pca}

# Create a DataFrame with the data to plot the Elbow curve
elbow_inertia_pca_df = pd.DataFrame(elbow_inertia_pca)
elbow_inertia_pca_df = elbow_inertia_pca_df.set_index('k')
elbow_inertia_pca_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_inertia_pca_plot = elbow_inertia_pca_df.hvplot.line(
        x="k",
        y="inertia_pca", 
)
elbow_inertia_pca_plot
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The best value for 'k' is 4. 


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No, the data shown for best k value does not differ from the original data. 

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the PCA data
model.fit(pca_data)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
cluster_predict = model.predict(pca_data)
# Print the resulting array of cluster values.
cluster_predict
```

```python
# Create a copy of the DataFrame with the PCA data
pca_cluster_predict = pca_data.copy()

# Add a new column to the DataFrame with the predicted clusters
pca_cluster_predict["clusters"] = cluster_predict

# Display sample data
pca_cluster_predict
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
cluster_scatter_pca = pca_cluster_predict.hvplot.scatter(
    x= "PC1", 
    y= "PC2", 
    by= "clusters", 
    hover_cols=["coin_id"],
)
cluster_scatter_pca
```
